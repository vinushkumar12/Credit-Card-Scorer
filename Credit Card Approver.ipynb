```python
from sklearn.model_selection import train_test_split
credits = credits.drop(["ZipCode", "DriversLicense"], axis=1)
```

```python
credits_train, credits_test = train_test_split(credits, test_size = 0.33, random_state = 42)
```

```python
credits_train = credits_train.replace('?', np.NaN)
credits_test = credits_test.replace('?', np.NaN)
credits_train.fillna(credits_train.mean(), inplace=True)
credits_test.fillna(credits_train.mean(), inplace=True)
```

```python
print(credits_train)
```

```python
for col in credits_train.columns:
    credits_train[col].fillna(credits_train[col].mode()[0], inplace = True)
    credits_test[col].fillna(credits_train[col].mode()[0], inplace = True)
```

```python
print(credits_train)
```

```python
credits_train = pd.get_dummies(credits_train)
credits_test = pd.get_dummies(credits_test)
credits_test = credits_test.reindex(columns=credits_train.columns, fill_value=0)
print(credits_test)
```
